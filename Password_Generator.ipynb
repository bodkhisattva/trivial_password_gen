```python
import random
from random import randint
import string

password_list = []

def space(password_list):
        num_space = input('Вам нужны пробелы в пароле?')
        if num_space.lower() in ('yes', 'да', 'Да', 'Yes'):
            final_passwords = []
            for element in password_list:
                if len(element) <= 5:
                    spaces1 = int(1)
                    your_password1 = element + ' ' * spaces1
                    your_final_password1 = ''.join(random.sample(your_password1, len(your_password1))) 
                    final_passwords.append(your_final_password1)
                elif 6<=len(element)<=10:
                    spaces2 = int(2)
                    your_password2 = element + ' ' * spaces2
                    your_final_password2 = ''.join(random.sample(your_password2, len(your_password2))) 
                    final_passwords.append(your_final_password2)
                elif 11<=len(element)<=19:
                    spaces3 = int(4)
                    your_password3 = element + ' ' * spaces3
                    your_final_password3 = ''.join(random.sample(your_password3, len(your_password3)))
                    final_passwords.append(your_final_password3)
                else:
                    spaces4 = int(6)
                    your_password4 = element + ' ' * spaces4
                    your_final_password4 = ''.join(random.sample(your_password4, len(your_password4))) 
                    final_passwords.append(your_final_password4)
            return f"Запишите то, что находится внутри квадратных скобок и одинарных кавычек{final_passwords}"
        else:
            if num_space.lower() in ('no', 'No', 'нет', 'Нет'):
                print('Пароль без пробелов.')  

user_input = input('Вы хотите сгенерировать пароль?')
if user_input.lower() in ('yes', 'да', 'Да', 'Yes'):
    number_pswd = int(input('Какой длины пароль вы хотите получить (3 символа минимально)?'))
    bigl = input('Вам нужны только большие/маленькие/большие и маленькие буквы?')
    if bigl.lower() in ('только большие', 'большие', 'Только большие', 'Большие', 'big', 'Big'):
        special = input('Вам нужны специальные символы?')
        if special.lower() in ('yes', 'да', 'Да', 'Yes'):
            big_spec = string.ascii_letters + string.ascii_uppercase + string.digits + string.hexdigits + string.octdigits + string.punctuation + string.printable
            passwd1 = ''.join(random.choice(big_spec) for i in range(number_pswd))
            password_list.append(passwd1)
            print(space(password_list))
        elif special.lower() in ('no', 'No', 'нет', 'Нет'):
            big_no_spec = string.ascii_letters + string.ascii_uppercase 
            passwd2 = ''.join(random.choice(big_no_spec) for i in range(number_pswd))
            password_list.append(passwd2)
            print(space(password_list))
    elif bigl.lower() in ('только маленькие', 'маленькие', 'Только маленькие', 'Маленькие', 'small', 'Small'):
        special1 = input('Вам нужны специальные символы?')
        if special1.lower() in ('yes', 'да', 'Да', 'Yes'):
            smal_spec = string.ascii_letters + string.ascii_lowercase + string.digits + string.hexdigits + string.octdigits + string.punctuation + string.printable 
            passwd3 = ''.join(random.choice(smal_spec) for i in range(number_pswd))
            password_list.append(passwd3)       
            print(space(password_list))
        elif special1.lower() in ('no', 'No', 'нет', 'Нет'):
            small_no_spec = string.ascii_letters + string.ascii_lowercase 
            passwd4 = ''.join(random.choice(small_no_spec) for i in range(number_pswd))
            password_list.append(passwd4)       
            print(space(password_list))
    elif bigl.lower() in ('Я хочу большие и маленькие буквы', 'Большие и маленькие', 'большие и маленькие', 'маленькие и большие', 'Маленькие и большие',
                         'хочу большие и маленькие', 'Хочу большие и маленькие', 'Большие и маленькие буквы', 'большие и маленькие буквы',
                         'Маленькие и большие буквы', 'маленькие и большие буквы'):  
        special2 = input('Вам нужны специальные символы?')
        if special2.lower() in ('yes', 'да', 'Да', 'Yes'):
            both_spec = string.ascii_letters + string.ascii_uppercase + string.ascii_lowercase + string.digits + string.hexdigits + string.octdigits + string.punctuation + string.printable
            passwd5 = ''.join(random.choice(both_spec) for i in range(number_pswd))
            password_list.append(passwd5)
            print(space(password_list))
        elif special2.lower() in ('no', 'No', 'нет', 'Нет'):
            both_no_spec = string.ascii_letters + string.ascii_lowercase + string.ascii_uppercase
            passwd6 = ''.join(random.choice(both_no_spec) for i in range(number_pswd))
            password_list.append(passwd6)       
            print(space(password_list))
                
                                                
elif user_input.lower() in ('no', 'No', 'нет', 'Нет'):
    phrase_import = input('Вы хотите сгенерировать парольную фразу?')
    if phrase_import.lower() in ('yes', 'да', 'Да', 'Yes'):
        number_phrase = int(input('Сколько слов вы хотите получить?'))
        with open('engwords.txt', 'r') as file:
            words = file.readlines()
            phrase_pswd = ''.join(random.choice(words) for _ in range(number_phrase))
            phrase_pswd = phrase_pswd.split()
            result_phrase = ' '.join(phrase_pswd)
            print(result_phrase)
else:
    raise ValueError('Введите другой ответ')
```
